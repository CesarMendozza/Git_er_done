```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd, numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

#### find jobs  by title summary , salary

```python
jobs = pd.read_csv("datasets/indeed_job_postings/indeed-scraped-job-postings.csv")
```

```python
vect = CountVectorizer(stop_words='english')
X_text = vect.fit_transform(jobs['summary'])
df = pd.DataFrame(X_text.toarray(), columns=vect.get_feature_names())
```

```python
df = pd.DataFrame(X_text.toarray(), columns=vect.get_feature_names())
df.sum().sort_values(ascending=False).head(15).sort_values().plot(kind="barh")
```

```python
jobs['title'].value_counts().head(10)
```

```python
#searching for data scientist jobs
data_mask = jobs['title'].str.contains('Data Scientist')
jobs[data_mask]
```

```python
#mapping 
jobs['data_scientist'] = jobs['title'].map(lambda title: 1 if "Data Scientist" in title else 0)
jobs
```

```python
#reindex
df.index= jobs['data_scientist']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(df, df.index, test_size=0.33, random_state=42)


logreg = LogisticRegression()
model = logreg.fit(df, df.index)

model.score(X_train, y_train), model.score(X_test, y_test)
```

###### futre problem  _ how much are non data scientist jobs, using data scientist skills

```python
y_hat = model.predict(X_test)
print(classification_report(y_test, y_hat))
```

```python
coefs = pd.DataFrame(model.coef_[0], index=df.columns, columns=["coef"])
coefs['coef_abs'] = coefs['coef'].abs()
coefs.sort_values('coef_abs', ascending=False)
```
